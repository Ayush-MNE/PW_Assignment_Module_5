3. Using the sales data, generate a heatmap to visualize the variation in sales across different months and days.

```python
np.random.seed(20)
data = {
    'Month': np.random.choice(['Jan', 'Feb', 'Mar', 'Apr', 'May'], 100),
    'Day': np.random.choice(range(1, 31), 100),
    'Sales': np.random.randint(1000, 5000, 100)
}
df = pd.DataFrame(data)
```

```python
# Create a pivot table to aggregate sales by Month and Day
sales_pivot = df.pivot_table(values='Sales', index='Day', columns='Month', aggfunc='mean')

# Create the heatmap using plotly express
fig = px.imshow(
    sales_pivot.values,labels=dict(x="Month", y="Day", color="Sales"),
    x=sales_pivot.columns,
    y=sales_pivot.index,
    color_continuous_scale='YlGnBu'
)

# Customize the layout
fig.update_layout(
    title='Heatmap of Sales Variation Across Months and Days'
)

# Show the plot
fig.show()
```

2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.

```python
import numpy as np
import pandas as pd
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.models import ColumnDataSource

# Enable Bokeh to render in the notebook (if in Jupyter)
output_notebook()

# Generate random data
np.random.seed(42)
n = 100
x = np.random.rand(n) * 100
y = np.random.rand(n) * 100
sizes = np.random.rand(n) * 50  # Random sizes
colors = np.random.randint(50, 255, n)  # Random colors as grayscale values

# Create a DataFrame
data = {'x': x, 'y': y, 'sizes': sizes, 'colors': colors}
df = pd.DataFrame(data)

# Create a ColumnDataSource
source = ColumnDataSource(df)

# Create a scatter plot using Bokeh
p = figure(title="Bokeh Scatter Plot with Varying Sizes and Colors", x_axis_label='X', y_axis_label='Y', width=700, height=700)

# Plot circles with sizes and colors based on the 'sizes' and 'colors' columns
p.circle('x', 'y', size='sizes', color='colors', fill_alpha=0.6, 
         source=source, line_color="black")

# Show the plot
show(p)
```

3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.

```python
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
```

```python
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource

# Enable Bokeh to render in the notebook (if using Jupyter)
output_notebook()

# Prepare the data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]

# Create a DataFrame or ColumnDataSource (for Bokeh plotting)
source = ColumnDataSource(data=dict(fruits=fruits, counts=counts))

# Create the figure
p = figure(x_range=fruits, title="Fruit Counts", 
           toolbar_location=None, tools="", 
           y_axis_label="Counts", width=600, height=400)

# Create the bar chart
p.vbar(x='fruits', top='counts', width=0.6, source=source, 
       legend_field="fruits", line_color='white',
       fill_color=factor_cmap('fruits', palette=['#ff9999', '#66b3ff', '#99ff99', '#ffcc99'], factors=fruits))

# Customize the chart
p.y_range.start = 0
p.xgrid.grid_line_color = None
p.legend.orientation = "horizontal"
p.legend.location = "top_center"

# Show the chart
show(p)
```

4. Create a Bokeh histogram to visualize the distribution of the given data.

```python
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Enable Bokeh to render in the notebook (if using Jupyter)
output_notebook()

# Generate data
data_hist = np.random.randn(1000)

# Compute histogram with 30 bins
hist, edges = np.histogram(data_hist, bins=30)

# Create a figure
p = figure(title="Histogram of Random Data", 
           x_axis_label='Value', y_axis_label='Frequency', 
           width=700, height=400)

# Plot the histogram using the quad glyph
p.quad(top=hist, bottom=0, left=edges[:-1], right=edges[1:], 
       fill_color="skyblue", line_color="white", alpha=0.7)

# Customize the plot
p.y_range.start = 0
p.xgrid.grid_line_color = None

# Show the plot
show(p)
```

5. Create a Bokeh heatmap using the provided dataset.

```python
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)
```

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import linear_cmap
from bokeh.models import ColorBar, LinearColorMapper

# Enable Bokeh to render in the notebook (if using Jupyter)
output_notebook()

# Data for the heatmap
data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)

# Create a figure
p = figure(title="Bokeh Heatmap", 
           x_axis_label='X', y_axis_label='Y', 
           width=400, height=400, 
           tools="", toolbar_location=None)

# Define a color mapper
color_mapper = LinearColorMapper(palette="Viridis256", low=data_heatmap.min(), high=data_heatmap.max())

# Plot the heatmap
p.image(image=[data_heatmap], x=0, y=0, dw=1, dh=1, color_mapper=color_mapper)

# Add color bar
color_bar = ColorBar(color_mapper=color_mapper, location=(0, 0))
p.add_layout(color_bar, 'right')

# Customize the plot
p.grid.grid_line_width = 0.5

# Show the heatmap
show(p)
```
