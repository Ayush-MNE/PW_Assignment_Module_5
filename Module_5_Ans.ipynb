1. Discuss String Slicing and provide example

Ans. String slicing is a technique used in programming to extract a portion of a string. It involves specifying a range of indices to obtain a substring. The general syntax for string slicing in Python is string[start:end:step], where:

start is the index where the slice begins (inclusive).
end is the index where the slice ends (exclusive).
step determines the increment between each index (optional).

```python
#Basic Slicing:
text = "Hello, World!"
slice1 = text[0:5]   # 'Hello'
slice2 = text[7:12]  # 'World'

#Using Negative Indices:
text = "Hello, World!"
slice1 = text[-6:-1] # 'World'
```

3. Describe how to access, modify and delete elements in a list with example 

Ans. Accessing, modifying, and deleting elements in a list are fundamental operations in Python. Here's a detailed explanation with examples for each operation.

Accessing Elements in a List
You can access elements in a list using indexing and slicing:
Indexing: Access a single element using its index. The index starts from 0.
Slicing: Access a sublist by specifying a range of indices

Modifying Elements in a List
You can modify elements in a list by assigning a new value to a specific index or a slice.

Deleting Elements in a List
You can delete elements from a list using del, remove(), pop(), and clear() methods:
del: Delete an element or a slice from the list.
remove(): Remove the first occurrence of a specific value.
pop(): Remove and return an element at a specific index.
clear(): Remove all elements from the list.

```python
#Accessing Elements in a List
# Create a list
my_list = [1, 2, 3, 4]

# Accessing elements
print(my_list[0])  # Output: 1

# Modifying elements
my_list[1] = 5
print(my_list)  # Output: [1, 5, 3, 4]

# Adding elements
my_list.append(6)
print(my_list)  # Output: [1, 5, 3, 4, 6]

# Removing elements
my_list.remove(5)
print(my_list)  # Output: [1, 3, 4, 6]

```

4. Compare and contrast tuples and lists with example

Ans. Tuples and lists are both data structures in Python used to store collections of items. However, they have different properties and use cases. Here's a detailed comparison:

Similarities

Ordered Collections:
Both tuples and lists maintain the order of elements.
Elements can be accessed using indexing and slicing.

Heterogeneous Elements:
Both can store elements of different data types (integers, strings, objects, etc.).


Differences

Mutability:
Lists are mutable, meaning their elements can be changed, added, or removed.
Tuples are immutable, meaning once they are created, their elements cannot be changed.

Syntax:
Lists are defined using square brackets [].
Tuples are defined using parentheses ().

Performance:
Tuples generally have a smaller memory footprint and can be faster than lists when iterating, due to their immutability.
Lists are more flexible but come with additional overhead.

Methods:
Lists have a wide range of built-in methods (e.g., append(), remove(), extend(), pop(), clear(), sort(), reverse()).
Tuples have fewer methods (e.g., count(), index()).

Usage:
Lists are typically used when the collection of items is expected to change (e.g., dynamic data).
Tuples are used for fixed collections of items, such as coordinates, or when you want to ensure the data remains constant (e.g., keys in a dictionary). 

```python
# Define a list
my_list = [1, 2, 3, 4]

# Accessing elements
print(my_list[0])  # Output: 1

# Modifying elements
my_list[1] = 5
print(my_list)  # Output: [1, 5, 3, 4]

# Adding elements
my_list.append(6)
print(my_list)  # Output: [1, 5, 3, 4, 6]

# Removing elements
my_list.remove(5)
print(my_list)  # Output: [1, 3, 4, 6]
```

```python
# Define a tuple
my_tuple = (1, 2, 3, 4)

# Accessing elements
print(my_tuple[0])  # Output: 1

# Tuples are immutable; the following lines would raise an error
# my_tuple[1] = 5  # TypeError: 'tuple' object does not support item assignment

# Tuples do not have methods like append() or remove()

# Tuples can be used as keys in a dictionary because they are immutable
my_dict = {(1, 2): "a value"}
print(my_dict[(1, 2)])  # Output: "a value"
```

5. Describe key features of set and provide example of their use

Ans. Sets in Python are a collection data type that provide powerful functionalities for handling unique elements. Here are some key features of sets, along with examples of their use:

Key Features of Sets

Unordered Collection:
Sets are unordered, meaning they do not record the position of elements.

Unique Elements:
Sets automatically ensure that all elements are unique. Duplicate values are not allowed.

Mutable:
Sets are mutable, meaning elements can be added or removed.

Dynamic Size:
Sets can grow or shrink dynamically as elements are added or removed.

No Indexing or Slicing:
Since sets are unordered, elements cannot be accessed by indexing or slicing.

Efficient Membership Testing:
Sets provide highly efficient methods for checking whether an element is present in the set (in operator).

Built-in Set Operations:
Sets support various mathematical set operations like union, intersection, difference, and symmetric difference.

```python
# Creating a set
my_set = {1, 2, 3, 4}
print(my_set)  # Output: {1, 2, 3, 4}

# Adding an element
my_set.add(5)
print(my_set)  # Output: {1, 2, 3, 4, 5}

# Adding duplicate elements (no effect)
my_set.add(3)
print(my_set)  # Output: {1, 2, 3, 4, 5}

# Removing an element
my_set.remove(4)
print(my_set)  # Output: {1, 2, 3, 5}

# Using discard() to remove an element (doesn't raise an error if the element is not present)
my_set.discard(6)  # No error
print(my_set)  # Output: {1, 2, 3, 5}
```

6. Discuss the use cases of tuples and sets in Python programming.

Ans. Use Cases for Tuples

Immutable Collections:
Tuples are immutable, meaning once created, their elements cannot be changed. This makes them ideal for storing data that should remain constant, like configuration settings or fixed collections of items such as coordinates in a 3D space.

Function Arguments:
Tuples are commonly used to return multiple values from a function. They are also useful for passing a fixed collection of arguments to a function.

Dictionary Keys:
Because tuples are immutable, they can be used as keys in dictionaries. This is useful for representing composite keys, like grid coordinates or database primary keys.

Fixed Data Structure:
Tuples are suitable for representing fixed collections of items, such as days of the week, months of the year, or RGB color values.

Heterogeneous Data:
Tuples can store elements of different data types, making them useful for grouping related but different types of data, like a person's name, age, and occupation.



Use Cases for Sets

Unique Elements:
Sets are ideal for storing unique elements and automatically eliminating duplicates. This is useful for tasks like removing duplicates from a list.

Membership Testing:
Sets provide efficient membership testing, making them useful for checking the existence of an element within a collection, such as ensuring a user is not added to a group twice.

Set Operations:
Sets support mathematical operations like union, intersection, difference, and symmetric difference, making them useful for tasks that involve comparing and combining collections of data, such as finding common or distinct items between datasets.

Data Cleaning:
Sets are useful in data cleaning tasks to find and retain only unique values in a dataset, such as extracting unique words from a large text.

Avoiding Duplicate Entries:
Sets are ideal for scenarios where duplicate entries must be avoided, such as maintaining a collection of unique user IDs, product codes, or inventory items.


Summary
Tuples are best used for immutable collections, returning multiple values from functions, keys in dictionaries, fixed data structures, and grouping heterogeneous data.
Sets are optimal for ensuring unique elements, efficient membership testing, performing set operations, data cleaning, and avoiding duplicate entries.
Understanding when to use tuples and sets helps in writing clear, efficient, and error-free code.

7. Describe how to add, modify and delete items in dictionary with example.

Ans. In Python, dictionaries are mutable data structures that allow you to store and manage data in key-value pairs. Here’s how you can add, modify, and delete items in a dictionary:

Adding Items to a Dictionary
Adding a New Key-Value Pair:
You can add a new key-value pair by assigning a value to a new key.

Modifying Items in a Dictionary
Modifying an Existing Key-Value Pair:
You can modify an existing value by reassigning a new value to an existing key.

Deleting Items from a Dictionary
Using del Keyword:
You can delete a key-value pair using the del keyword.
Using pop() Method:
You can remove a key-value pair and get its value using the pop() method.
Using popitem() Method:
You can remove and return the last inserted key-value pair using the popitem() method.
Using clear() Method:
You can remove all items from the dictionary using the clear() method.

```python
# Creating an empty dictionary
my_dict = {}

# Adding items
my_dict['name'] = 'Alice'
my_dict['age'] = 25
my_dict['city'] = 'New York'

# Modifying items
my_dict['age'] = 26  # Modify age
my_dict['city'] = 'Los Angeles'  # Modify city

# Deleting items
del my_dict['city']  # Delete using del
age = my_dict.pop('age')  # Delete using pop and store the value in variable 'age'
last_item = my_dict.popitem()  # Delete the last inserted item using popitem

# Clearing all items
my_dict.clear()  # Clear the dictionary

# Displaying the dictionary at each step
print("After adding items: ", {'name': 'Alice', 'age': 25, 'city': 'New York'})
print("After modifying items: ", {'name': 'Alice', 'age': 26, 'city': 'Los Angeles'})
print("After deleting 'city' using del: ", {'name': 'Alice', 'age': 26})
print("Age value removed using pop: ", age)  # Output: 26
print("Last item removed using popitem: ", last_item)  # Output: ('name', 'Alice')
print("After clearing the dictionary: ", my_dict)  # Output: {}
```

8. Discuss the importance of dictionary keys being immutable and provide example.

Ans. In Python, dictionary keys must be immutable. This requirement ensures that the keys are hashable, meaning that they have a fixed hash value for their entire lifetime, allowing for efficient retrieval. The immutability of dictionary keys is crucial for maintaining the integrity and performance of the dictionary. Here’s why immutable keys are important and an example to illustrate this concept:

Reasons for Immutable Keys

Consistency:
Immutable keys ensure that the key's hash value remains constant throughout its lifetime. This consistency is vital for the internal hash table mechanism used by dictionaries to store and retrieve key-value pairs efficiently.

Efficiency:
The hash table relies on the hash values of keys to index and look up values quickly. If keys were mutable, their hash values could change, leading to potential mismatches and inefficient lookups.

Integrity:
Using immutable keys prevents accidental changes to keys that could corrupt the dictionary's structure. This immutability guarantees the integrity of the data stored in the dictionary.

Hashing:
Immutable keys are hashable, which means they can be used as indices in the hash table. Hashable objects must have a constant hash value that does not change during their lifetime.

```python
# Example with an immutable key (tuple)
coordinates = (10, 20)
location_dict = {coordinates: "Home"}

# Trying to change the coordinates
# This will fail because tuples are immutable
try:
    coordinates[0] = 15
except TypeError as e:
    print(f"Error: {e}")

# Example with a mutable key (list) - this will cause an error
try:
    mutable_key = [10, 20]
    location_dict = {mutable_key: "Office"}  # This will raise a TypeError
except TypeError as e:
    print(f"Error: {e}")

# Demonstrating dictionary lookup with immutable keys
coordinates = (10, 20)
location_dict = {coordinates: "Home"}
print(location_dict[(10, 20)])  # Output: Home
```
