3. Explain the following terms
   1. The difference in np.array, np.asarray and np.asanyarray
   2. The difference between Deep copy and shallow copy.

1. Difference between np.array, np.asarray, and np.asanyarray:

np.array:
Creates a new NumPy array.
If the input is already an array, it will create a copy unless copy=False is specified.
It's more flexible, as you can specify the dtype, order, etc.

np.asarray:
Converts input to a NumPy array without copying if the input is already an array.
If the input is not a NumPy array (e.g., a Python list), it will convert it to one.

np.asanyarray:
Similar to np.asarray, but if the input is a subclass of np.ndarray (like a masked array), it will return the subclass instead of converting it to a base ndarray.

2. Difference between Deep Copy and Shallow Copy:

Shallow Copy:
Creates a new object, but does not recursively copy nested objects. Instead, it copies references to the original objects.
Changes in the nested objects (like lists or dictionaries inside the original object) will affect both the original and the shallow copy.

Deep Copy:
Creates a new object and recursively copies all nested objects. It copies everything, ensuring that the copied object is completely independent of the original.

17. Create a Seaborn scatter plot of two random arrays, color points based on their position relative to the
origin (quadrants), add a legend, label the axes, and set the title as 'Quadrant-wise Scatter Plot'.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

# Generate two random arrays
np.random.seed(42)  # For reproducibility
x = np.random.randn(100)
y = np.random.randn(100)

# Determine the quadrants based on position relative to the origin
quadrants = []
for i in range(len(x)):
    if x[i] >= 0 and y[i] >= 0:
        quadrants.append('Quadrant I')
    elif x[i] < 0 and y[i] >= 0:
        quadrants.append('Quadrant II')
    elif x[i] < 0 and y[i] < 0:
        quadrants.append('Quadrant III')
    else:
        quadrants.append('Quadrant IV')

# Create a scatter plot
plt.figure(figsize=(8, 6))
scatter_plot = sns.scatterplot(x=x, y=y, hue=quadrants, palette='Set1')

# Add quadrant lines (x=0 and y=0)
plt.axhline(0, color='black', linestyle='--', linewidth=1)
plt.axvline(0, color='black', linestyle='--', linewidth=1)

# Customize the plot
plt.title('Quadrant-wise Scatter Plot')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend(title='Quadrants', loc='upper right')

# Display the plot
plt.show()
```

18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.

```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook

# Enable Bokeh output in the notebook
output_notebook()

# Generate the x values and the sine wave
x = np.linspace(0, 4 * np.pi, 500)
y = np.sin(x)

# Create a new plot
p = figure(title="Sine Wave Function", width=800, height=400)

# Add a line renderer with the sine wave
p.line(x, y, line_width=2, color='blue', legend_label="Sine")

# Add grid lines, labels, and title
p.xgrid.grid_line_color = 'gray'
p.ygrid.grid_line_color = 'gray'
p.xaxis.axis_label = "X-axis"
p.yaxis.axis_label = "Y-axis"
p.legend.location = "top_left"

# Show the plot
show(p)
```

19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their
values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical
Bar Chart'.


```python
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
from bokeh.transform import factor_cmap
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.palettes import Viridis256

# Enable Bokeh output in the notebook
output_notebook()

# Generate random categorical data
categories = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(1, 100, size=len(categories))

# Create a data source
source = ColumnDataSource(data=dict(categories=categories, values=values))

# Create a new figure for the bar chart
p = figure(x_range=categories, height=400, width=600, title="Random Categorical Bar Chart",
           toolbar_location=None, tools="")

# Add bars with color based on values
p.vbar(x='categories', top='values', width=0.9, source=source, 
       fill_color=factor_cmap('categories', palette=Viridis256[:len(categories)], factors=categories), 
       line_color='black')

# Add hover tooltips to display the exact values
hover = HoverTool()
hover.tooltips = [("Category", "@categories"), ("Value", "@values")]
p.add_tools(hover)

# Customize the chart
p.xgrid.grid_line_color = None
p.y_range.start = 0
p.xaxis.axis_label = "Categories"
p.yaxis.axis_label = "Values"

# Show the plot
show(p)
```

20. Using Plotly, create a basic line plot of a randomly generated dataset, label the axes, and set the title as
'Simple Line Plot'.

```python
import numpy as np
import plotly.graph_objects as go

# Generate random data
np.random.seed(42)  # For reproducibility
x = np.arange(0, 100)
y = np.random.randn(100).cumsum()

# Create a basic line plot
fig = go.Figure()

fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Random Data'))

# Customize the plot with labels and title
fig.update_layout(
    title='Simple Line Plot',
    xaxis_title='X-axis',
    yaxis_title='Y-axis'
)

# Show the plot
fig.show()
```

21. Using Plotly, create an interactive pie chart of randomly generated data, add labels and percentages, set
the title as 'Interactive Pie Chart'.

```python
import numpy as np
import plotly.graph_objects as go

# Generate random data
np.random.seed(42)  # For reproducibility
labels = ['Category A', 'Category B', 'Category C', 'Category D', 'Category E']
values = np.random.randint(10, 100, size=len(labels))

# Create an interactive pie chart
fig = go.Figure(data=[go.Pie(labels=labels, values=values, hole=0.0, 
                             hoverinfo='label+percent', textinfo='label+percent')])

# Customize the plot with a title
fig.update_layout(title_text='Interactive Pie Chart')

# Show the plot
fig.show()
```
